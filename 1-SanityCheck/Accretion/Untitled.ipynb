# Convergence:

We suppose that the accreted mass we measure is a combination of the real accreted mass plus a contribution of numerical accreted mass:

$$
M_{\rm measured} = M_{\rm real} + M_{\rm num}.
$$

We also supose that the numerical accreted mass tends to go to zero as we increase the resolution, in such a way that:

$$
M_{\rm measured} = M_{\rm real} \left( 1 + C_0 N_{\rm part}^{-n} \right).
$$

We would like $n > 1$. If things work fine, that should be the case. We thus measured several accreted masses for several resolutions to find $C_0$ and $n$. For two different resolutions, with $N_2 > N_1$, we define:

$$
f_{21} = M_{\rm measured} (N_2) / M_{\rm measured}(N_1).
$$

We can then find $C0$. After some algebra:

$$
C_0 = \frac{1-f_{21}}{f_{21} - (\frac{N1}{N2})^{n}} N_1^{n}.
$$

Ideally, $C_0$ should be the same for any couple $N_1, N_2$, so we can use that expression for different couples of resolutions such that $N_2 = 2 N_1$ and find $n$ solving a quadratic equation. We solve the quadratic equation for different couples $(N_1,N_2)$ and we should get the same roots for $n$... ideally. Numerically, however, we don't find exactly the same roots and we have to tweak the numbers a bit. That's the purpose of this ipython notebook. By tweaking the numbers and using the blackboard a lot, we find that a compatible couple of $C_0$ and $n$ is given by:

$$
C_0 = 10^6 
$$

$$
n = 2
$$

```python

x = np.linspace(1.9,2.25,1025)
A1 = 2**x * (0.39) / ((2**x)*0.61-1)
A2 = 1**x * (0.64) / ((2**x)*0.4-1)
A3 = 0.5**x*(0.67) / ((2**x)*0.29-1)
A0 = x**0
pl.plot(x,A1,'r-')
pl.plot(x,A2,'g-')
pl.plot(x,A3,'b-')
pl.plot(x,A0,'k--')
pl.xlabel("$n$")
pl.ylabel("$C_0 [\times 10^6]$")
pl.grid()
pl.show()
```
